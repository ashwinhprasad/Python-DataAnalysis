```python
#importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

# SubPlots

```python
#basic subplots
x = np.arange(11)
y = x**2
plt.subplot(2,1,1)
plt.plot(x,y,color='red',label="y=x**2")
plt.legend()
plt.subplot(2,1,2)
plt.plot(y,x,linestyle='--',label="x = y**2")
plt.legend()
plt.show()
```

```python
#object oriented approach
fig = plt.figure(figsize=(8,4))
# firs plot
ax1 = fig.add_axes([0.1,0.1,0.45,1])
bars = ax1.bar(['A','B','C'],[5,2,4],color=['orange','grey','green'])
hatches = ['o','/','x']
for bar in bars:
    bar.set_hatch(hatches.pop(0))
    
# second plot
ax2 = fig.add_axes([0.65,0.1,0.45,1])
ax2.plot(x,y,linewidth=2,linestyle='--',color='black',marker='x',markersize='10',label='Random Line')
ax2.legend()
plt.savefig('./plots/subplot-objects',dpi=300)
```

```python
#default way
fig, axes = plt.subplots(2,1)
axes[0].pie(x=[30,50,20],labels=['Anime','Sleep','Eat'])
axes[1].scatter(x,y)
plt.show()
```
