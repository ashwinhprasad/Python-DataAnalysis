```python
#importing the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

## Basic Plots

```python
x = [1,2,3,4]
y = [2,3,3,5]
plt.figure(figsize=(5,3),dpi=100)
plt.style.use('fivethirtyeight')
plt.title("Basic Graph",fontdict={"fontsize":20})
plt.xlabel("X - Axis")
plt.ylabel("Y - Axis")
plt.xticks([0,1,2,3,4])
plt.yticks([2,3,4,5])
plt.plot(np.arange(1,5),np.arange(1,5)**2,color='green',linewidth=2,marker='^',markeredgecolor='red',markersize=8,label='another random line')
plt.plot(x,y,label="random line",linewidth=2,color='red',marker='.',markersize=15,markeredgecolor='blue',linestyle='--')
plt.legend()
plt.savefig('./plots/plot1.png',dpi=300)
```

## Bar Plot

```python
labels = ['A','B','C']
values =  [200,300,100]
plt.figure(figsize=(4,2.5),dpi=150)
bars = plt.bar(labels,values,color=['red','blue','green'])
hatches = ['/','*','o']
for bar in bars:
    bar.set_hatch(hatches.pop(0))
plt.savefig('./plots/barplot.png',dpi=300)
```

## Scatter Plot

```python
x = [0,1,2,3,4,5,6,7]
y = [2,4,5,1,3,5,8,1]
plt.figure(figsize=(5,3))
plt.style.use("default")
plt.scatter(x,y,marker='x')
plt.savefig("./plots/scatter.png")
```
